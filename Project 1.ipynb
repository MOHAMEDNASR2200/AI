```python
import seaborn as sns
import numpy as np
import pandas as pd
import statistics as st 
import matplotlib.pyplot as plt
data=pd.read_csv(r"C:\Users\Sara\Downloads\insurance-data.csv")
#understand data
print(data.describe())
print(data.info())
print(data["smoker"].unique())
print(data["region"].unique()
n_data=data.loc[:, ~data.columns.isin(['sex', 'region','smoker'])]
sns.pairplot(n_data)
#visualize the data
plt.boxplot(n_data,labels=n_data.columns)
plt.show()
#There are outliers in bmi and charges, however they sound logical
plt.boxplot(n_data["bmi"])
plt.show()
plt.boxplot(n_data["charges"])
plt.show()
#no cleaning required
#stan
def stan(A,name):
    R=list(A[name])
    s=0
    counter=0
    for i in R:
       s=s+i
       counter=counter+1
    M=s/counter
    v=0
    for j in R:
        v=v+((j-M)**2)
    t=(v/counter)**0.5
    A[name]=(A[name]-M)/t
    return A[name]
lis=list(n_data.columns)
for i in lis:
    data[i]=stan(data,i)
#Encode categorical data ["smoker","region",sex]
#label encoding for smoker and sex
data["smoker"].loc[data["smoker"]=="yes"]=1
data["smoker"].loc[data["smoker"]=="no"]=0
data["sex"].loc[data["sex"]=="male"]=1
data["sex"].loc[data["sex"]=="female"]=0
#one-hot encoding for region
l=list(data["region"].unique())
for z in l:
    n=list(data["region"]==z)
    for i in range(len(n)):
        if n[i]==True:
            n[i]=1
        else:
            n[i]=0
    data.loc[:, z] = n
data.drop(columns="region",inplace=True)
data=data.astype({"sex":int,"smoker":int})
print(data.corr()["charges"])
```
